```python
train['Mileage']=train["Mileage"].astype(float)
train['Engine']=train["Engine"].astype(float)
train['Power'] = train['Power'].replace('null', np.nan)
train["Power"]=train["Power"].astype(float)
```

###### Drop New_price since the value present in this col is more than 50% value.

###### DROP Name , Location

```python
## DROP Name,Location,New_Price
new_train=train.drop(["Name","Location","New_Price"],axis=1).copy()
```

```python
new_train.columns
```

```python
features=new_train.drop("Price",axis=1)
label=new_train["Price"]
```

```python
num_encoder=["Mileage","Engine","Power","Seats"]
num=['Year', 'Kilometers_Driven']
cat=['Fuel_Type', 'Transmission', 'Owner_Type']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.compose import ColumnTransformer,make_column_transformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler,FunctionTransformer,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
pipe_num_val=make_pipeline(
    StandardScaler()
)

pipe_cat_val=make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder()
)
pipe_num_val_imputer=make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
)
```

```python
inner=ColumnTransformer([
    ("Num_val",pipe_num_val,num),
    ("cat_val",pipe_cat_val,cat),
    ("Num_val_imputer",pipe_num_val_imputer,num_encoder)
])
```

```python
pipeline=Pipeline([
    ("Pipeline",inner),
    ("model",RandomForestRegressor())
])
pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,label, test_size=0.2, random_state=42)
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_hat=pipeline.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
acc=r2_score(y_train,y_hat)*100
print("ACCURACY:"," "*10,acc)

xm=mean_squared_error(y_train,y_hat)
error=np.sqrt(xm)
print("ERROR:"," "*13,error)

error_per=(error/y_train.mean())*100
print("ERROR PERCENTAGE:","  ",error_per)
```

```python
y_hat_test=pipeline.predict(X_test)
```

```python
acc=r2_score(y_test,y_hat_test)*100
print("ACCURACY:"," "*10,acc)

xm=mean_squared_error(y_test,y_hat_test)
error=np.sqrt(xm)
print("ERROR:"," "*13,error)

error_per=(error/y_test.mean())*100
print("ERROR PERCENTAGE:","  ",error_per)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
param_grid = {
    'model__n_estimators': [50,100,140,170],
}

grid=GridSearchCV(pipeline,param_grid,cv=3, scoring='neg_mean_squared_error',return_train_score=True)
grid
```

```python
grid.fit(X_train,y_train)
```
